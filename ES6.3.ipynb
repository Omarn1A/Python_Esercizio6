```python
from sqlalchemy import create_engine
import dotenv
import os
import pandas as pd
dotenv.load_dotenv(override=True)
username = os.getenv('username')
password = os.getenv('password')
host = os.getenv('host')
dbname = os.getenv('dbname')

import pandas as pd
dbname = "AdventureWorksDW"
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
query = """SELECT *
        FROM dimcustomer
        """
df = pd.read_sql(query, db_engine)
```

```python
df['FirstName'].str.lower()
```

```python
df['LastName'].str.upper()
```

```python
df[['Username', 'Domain']] = df['EmailAddress'].str.split('@', expand=True)
df
```

```python
df['PhoneParts'] = df['Phone'].str.findall(r'\d+|[^0-9\s]+')
df
```

```python
contains_21 = df[df['EmailAddress'].str.contains('21')]
print(contains_21)
```

```python
contains_20_or_10 = df[df['EmailAddress'].str.contains('20|10')]
print(contains_20_or_10)
```

```python
df['EmailLength'] = df['EmailAddress'].str.len()
df
```

```python
ten_longest_emails = df.nlargest(10, 'EmailLength')
ten_shortest_emails = df.nsmallest(10, 'EmailLength')
print(ten_longest_emails)
print(ten_shortest_emails)
```

```python
df['EmailAddress'] = df['EmailAddress'].str.replace('adventure-works.com', 'aw-db.com')
df
```

```python
contains_street = df[df['AddressLine1'].str.contains('Street', case=False)]
print(contains_street)
```
