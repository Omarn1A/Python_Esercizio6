```python
df['status_published_convert']=pd.to_datetime(df['status_published'])
df
```

```python
posts_2012 = df[(df['status_published_convert'].dt.year == 2012)]
posts_2012
```

```python
posts_may_2018 = df[(df['status_published_convert'].dt.year == 2018) & (df['status_published_convert'].dt.month == 5)]
posts_may_2018
```

```python
weekend_posts = df[df['status_published_convert'].dt.dayofweek.isin([5, 6])]
weekday_posts = df[~df['status_published_convert'].dt.dayofweek.isin([5, 6])]
weekday_posts
```

```python
first_post_year = df.groupby(df['status_published_convert'].dt.year)['status_published_convert'].min()
last_post_year = df.groupby(df['status_published_convert'].dt.year)['status_published_convert'].max()
```

```python
df['status_published_with_timezone'] = df['status_published_convert'].dt.tz_localize('UTC').dt.tz_convert('Europe/Rome')
df
```

```python
df['time_diff'] = df['status_published_convert'].diff()
df
```

```python
num_post_types = df['status_published_convert'].nunique()
print(num_post_types)
```

```python
post_types_counts = df['status_published_convert'].value_counts()
print(post_types_counts)
```
