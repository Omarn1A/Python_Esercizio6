```python
import pandas as pd
df=pd.read_csv('automobile.csv')
```

```python
null_counts = df.isnull().sum()
# Stampiamo i conteggi dei valori nulli per ogni colonna
print("Valori nulli per ogni colonna:")
print(null_counts)
```

```python
null_doors = df[df['num-of-doors'].isnull()]

# Stampiamo le righe con valori nulli nella colonna 'num-of-doors'
print("\nRighe con valori nulli nella colonna 'num-of-doors':")
print(null_doors)
```

```python
# Esaminiamo i dati per trovare valori da sostituire ai valori nulli in 'num-of-doors'
# Vediamo quante porte hanno le auto con lo stesso make e body-style
replacement_values = df.groupby(['make', 'body-style'])['num-of-doors'].agg(lambda x:x.value_counts().index[0]).reset_index()

# Uniamo i valori sostitutivi al DataFrame originale
df_filled = pd.merge(df, replacement_values, on=['make', 'body-style'], how='left', suffixes=('', '_replacement'))

# Sostituiamo i valori nulli nella colonna 'num-of-doors' con i valori calcolati
df_filled['num-of-doors'].fillna(df_filled['num-of-doors_replacement'], inplace=True)

# Ora verifichiamo se ci sono ancora valori nulli dopo la sostituzione
null_counts_after = df_filled.isnull().sum()

# Stampiamo i conteggi dei valori nulli per ogni colonna dopo la sostituzione
print("\nValori nulli per ogni colonna dopo la sostituzione:")
print(null_counts_after)

# Stampiamo le prime righe del DataFrame con i valori sostituiti
print("\nPrimi 10 righe del DataFrame con valori sostituiti:")
print(df_filled.head(10))
```
